# visualizing using countplot 

```python
sns.set(style ='darkgrid')
fi, axs = plt.subplots(1,2, figsize = (20,8))

sns.countplot(x = 'gender', data = df , ax = axs[0])
sns.countplot(x = 'race/ethnicity', data = df, ax = axs[1])
```

# Histogram plot

```python
sns.set(style = 'darkgrid')
plt.hist(df['math score'], 30, color = '#F08080')
plt.title= ("Math Score")
plt.show()

sns.set(style = 'darkgrid')
plt.hist(df['reading score'], 30, color = '#008080')
plt.title=('Reading Score')
plt.show()
```

```python
sns.set(style = 'darkgrid')
plt.hist(df['writing score'], 30, color = '#FFBE33')
plt.title= ("writing score")
plt.show()
```
